```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
with open('data/minst.pkl', 'rb') as f:
    data = pickle.load(f)
X_train, y_train = data['X_train'], data['y_train']
X_test, y_test = data['X_test'], data['y_test']
```

```python
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5): 
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
```

```python
# Q2

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("X_train[i]:", X_train[0].shape)
print("X_test[i]:", X_test[0].shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
# Q3

N_train = X_train.shape[0]
N_test = X_test.shape[0]

X_train_reshape = X_train.reshape(N_train, 28*28)
X_test_reshape = X_test.reshape(N_test, 28*28)

print(X_train_reshape.shape)
print(X_test_reshape.shape)
print(X_test[0].reshape(1, 784).shape)
```

```python
#Q4 Used AI to make sure I computed this right
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

Xtr = X_train_reshape.astype(np.float32) / 255.0
Xte = X_test_reshape.astype(np.float32) / 255.0

Xtr_sub, Xval, ytr_sub, yval = train_test_split(
    Xtr, y_train, test_size=10000, random_state=0, stratify=y_train
)

ks = range(1, 26, 2)  # odd k avoids ties
scores = {
    k: accuracy_score(yval, KNeighborsClassifier(n_neighbors=k).fit(Xtr_sub, ytr_sub).predict(Xval))
    for k in ks
}
best_k = max(scores, key=scores.get)

print("k grid:", list(ks))
print("val accuracies:", [round(scores[k], 4) for k in ks])
print("best k =", best_k, "with val acc =", round(scores[best_k], 4))

#The optimal k is 1 which had the highest validation score
```

```python
#Q5 Used some AI help with this complex task(s)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# scale and refit on all training data
Xtr_full = X_train_reshape.astype(np.float32) / 255.0
Xte_full = X_test_reshape.astype(np.float32) / 255.0

knn = KNeighborsClassifier(n_neighbors=best_k, n_jobs=-1)
knn.fit(Xtr_full, y_train)

y_hat = knn.predict(Xte_full)
acc = accuracy_score(y_test, y_hat)
print(f"Test accuracy (k={best_k}): {acc:.4f}")

# confusion matrix
cm = confusion_matrix(y_test, y_hat)
disp = ConfusionMatrixDisplay(cm)
fig = plt.figure()
disp.plot(values_format="d", ax=plt.gca())
plt.title(f"kNN Confusion Matrix (k={best_k})")
plt.show()

#summary
cm_off = cm.copy()
np.fill_diagonal(cm_off, 0)
pairs = np.dstack(np.unravel_index(np.argsort(cm_off.ravel())[::-1], cm_off.shape))[0]
top = [(int(i), int(j), int(cm_off[i, j])) for i, j in pairs[:5] if cm_off[i, j] > 0]
print("Top confusions (truepred, count):", top)

#Answer the qualitative part with a write-up
#Using k-NN, I selected the optimal k = 1 via cross-validation on the training set.
#After refitting on all training data, the test accuracy is 96.91%. The confusion matrix indicates that the most common errors are 49 (22), 35 (19), 27 (16), 71 (14), and 83 (14).
#These confusions follow a clear pattern: pairs with similar strokes or shapes are hardest to separate
#The classifier performs very well, with errors concentrated among visually similar digits.
```
